Creating a basic countplot to get number of transaction for each product line and change the figure size

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
mart = pd.read_excel("../Data/supermarket_sales.xlsx")

plt.figure(figsize=(15,4))

sns.countplot(x = "Product line", data= mart)
plt.show()
mart.columns
```

Making it horizontal bar plot


```python
sns.countplot(y = "Product line", data= mart )
```

```python
plt.figure(figsize=(14,5))
sns.countplot(x = "Product line", data= mart, hue="Gender")
```

Seaborn barplot 
create a basic barplot() 

```python
plt.figure(figsize=(14,5))
sns.barplot(x = "Product line",y = "Total", data= mart)
```
